```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(".\\price_paid_records.csv")
data.columns = data.columns.str.lower()
```

```python
# Histogram of price distribution
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Limit the x-axis to focus on the main distribution
price_cap = data['price'].quantile(0.99)
filtered_prices = data[data['price'] <= price_cap]

# Plot with log scale and aesthetic improvements
sns.histplot(filtered_prices['price'], bins=100, kde=True, color="skyblue", edgecolor="black")
plt.xscale('log')  # Log-transform the x-axis for better visibility of lower prices

# Title and labels
plt.title('Log-Scaled Price Distribution (Capped at 99th Percentile)', fontsize=16)
plt.xlabel('Price (log scale)', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.show()
```

```python
# Box plot of price distribution
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Customizing the box plot
sns.boxplot(x='price', data=data, palette="Set2")
plt.xscale('log')  # Log-transform the x-axis for better visibility of lower prices

# Adding title and labels
plt.title('Log-Scaled Price Box Plot', fontsize=16)
plt.xlabel('Price (log scale)', fontsize=12)

# Show the plot
plt.show()
```

```python
# Density plot of price distribution
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# Plot with log scale and aesthetic improvements
sns.kdeplot(filtered_prices['price'], shade=True, color="skyblue")
plt.xscale('log')  # Log-transform the x-axis for better visibility of lower prices

# Adding title and labels
plt.title('Log-Scaled Price Density Plot', fontsize=16)
plt.xlabel('Price (log scale)', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Show the plot
plt.show()
```

```python
# Time series plot of price over time

data['date of transfer'] = pd.to_datetime(data['date of transfer'])
data.set_index('date of transfer').resample('M')['price'].mean().plot(figsize=(12, 6))
plt.title('Average Price Over Time')
plt.ylabel('Average Price')
plt.xlabel('Date')
plt.show()
# Plotting price trends for each district over time
plt.figure(figsize=(14, 8))

# Group by date and district, then calculate the mean price
district_prices = data.groupby([data['date of transfer'].dt.to_period('M'), 'district'])['price'].mean().unstack()

# Plot each district's price trend
for district in district_prices.columns:
    plt.plot(district_prices.index.to_timestamp(), district_prices[district], label=district)

# Adding title and labels
plt.title('Average Price Over Time by District')
plt.ylabel('Average Price')
plt.xlabel('Date')
plt.legend(title='District', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
# Heat map of price over time

data['year'] = data['date of transfer'].dt.year
data['month'] = data['date of transfer'].dt.month
pivot = data.pivot_table(values='price', index='year', columns='month', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Average Price by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

```python
# Box plot of price by property type
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Customizing the box plot with palette and order
sns.boxplot(x='property type', y='price', data=data, palette="Set3")
plt.yscale('log')  # Log-transform the y-axis for better visibility of lower prices

# Adding title and labels with improved font sizes
plt.title('Log-Scaled Price Distribution by Property Type', fontsize=16)
plt.xlabel('Property Type', fontsize=14)
plt.ylabel('Price (log scale)', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Bar plot of average price by property type
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Use a subset of the data for quicker loading
subset_data = data.sample(n=10000, random_state=1)

# Create the bar plot without confidence intervals
sns.barplot(x='property type', y='price', data=subset_data, ci=None, palette="Set2")

# Adding title and labels
plt.title('Average Price by Property Type', fontsize=16)
plt.xlabel('Property Type', fontsize=14)
plt.ylabel('Average Price', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
#Average price of top 10 towns/cities
top_towns = data['town/city'].value_counts().head(10).index
plt.figure(figsize=(12, 6))
sns.barplot(x='town/city', y='price', ci=None, data=data[data['town/city'].isin(top_towns)], estimator='mean')
plt.title('Average Price by Town/City')
plt.xlabel('Town/City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of price by new vs. old properties

plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Customizing the box plot with palette and order
sns.boxplot(x='old/new', y='price', data=data, palette="Set3")
plt.yscale('log')  # Log-transform the y-axis for better visibility of lower prices

# Adding title and labels with improved font sizes
plt.title('Log-Scaled Price Distribution by New vs. Old Properties', fontsize=16)
plt.xlabel('New or Old', fontsize=14)
plt.ylabel('Price (log scale)', fontsize=14)

# Show the plot
plt.show()
```
