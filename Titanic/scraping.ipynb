```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(".\\Titanic-Dataset.csv")
```

```python
#data['Age'] = data['Age'].fillna(data['Age'].median())
# Fill missing values in 'Embarked' column with the mode (most frequent value)
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data.columns = data.columns.str.lower()
```

```python
data
```

```python
#plot survived
plt.figure(figsize=(10, 6))
data['survived'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
#plot PCLass
plt.figure(figsize=(10, 6))
data['pclass'].value_counts().plot(kind='bar', color='blue')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Count')
plt.show()
```

```python
#plot sex
plt.figure(figsize=(10, 6))
data['sex'].value_counts().plot(kind='bar', color=['purple', 'orange'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
#plot age
plt.figure(figsize=(10, 6))
data['age'].plot(kind='hist', bins=30, color='teal', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#plot sibsp
plt.figure(figsize=(10, 6))
data['sibsp'].value_counts().plot(kind='bar', color='salmon')
plt.title('Number of Siblings/Spouses Aboard')
plt.xlabel('SibSp')
plt.ylabel('Count')
plt.show()
```

```python
#plot parch
plt.figure(figsize=(10, 6))
data['parch'].value_counts().plot(kind='bar', color='olive')
plt.title('Number of Parents/Children Aboard')
plt.xlabel('Parch')
plt.ylabel('Count')
plt.show()
```

```python
#plot fare
plt.figure(figsize=(10, 6))
data['fare'].plot(kind='hist', bins=30, color='cyan', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
#plot embarked
plt.figure(figsize=(10, 6))
data['embarked'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Port of Embarkation Distribution')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.xticks([0, 1, 2], ['Southampton (S)', 'Cherbourg (C)', 'Queenstown (Q)'])
plt.show()
```
